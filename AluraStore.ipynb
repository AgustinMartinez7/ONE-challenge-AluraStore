### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

#1. Análisis de facturación



```python
facturacion1 = tienda['Precio'].sum()
print(f'El ingreso total de la tienda 1 es de: {facturacion1} pesos.')
```

```python
facturacion2 = tienda2['Precio'].sum()
print(f'El ingreso total de la tienda 2 es de: {facturacion2} pesos.')
```

```python
facturacion3 = tienda3['Precio'].sum()
print(f'El ingreso total de la tienda 3 es de: {facturacion3} pesos.')
```

```python
facturacion4 = tienda4['Precio'].sum()
print(f'El ingreso total de la tienda 4 es de: {facturacion4} pesos.')
```

```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
facturacion = [facturacion1, facturacion2, facturacion3, facturacion4]

plt.figure(figsize=(10, 6))
plt.bar(tiendas, facturacion, color=['skyblue', 'salmon', 'lightgreen', 'orange'])

plt.title('Comparación de ingresos totales por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingreso total (Pesos)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

# 2. Ventas por categoría

```python
ventasCategoria1 = tienda.groupby('Categoría del Producto')['Categoría del Producto'].value_counts().sort_values(ascending=False)
ventasCategoria1
```

```python
ventasCategoria2 = tienda2.groupby('Categoría del Producto')['Categoría del Producto'].value_counts().sort_values(ascending=False)
ventasCategoria2
```

```python
ventasCategoria3 = tienda3.groupby('Categoría del Producto')['Categoría del Producto'].value_counts().sort_values(ascending=False)
ventasCategoria3
```

```python
ventasCategoria4 = tienda4.groupby('Categoría del Producto')['Categoría del Producto'].value_counts().sort_values(ascending=False)
ventasCategoria4
```

```python
import matplotlib.pyplot as plt
import pandas as pd

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()

titulos = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
datos_categorias = [ventasCategoria1, ventasCategoria2, ventasCategoria3, ventasCategoria4]

for i, data in enumerate(datos_categorias):
    valores = data.iloc[:, 0] if isinstance(data, pd.DataFrame) else data

    axes[i].pie(valores, labels=data.index, autopct='%1.1f%%', startangle=140,
                shadow=True, explode=[0.05] + [0]*(len(data)-1))
    axes[i].set_title(f'Distribución de ventas - {titulos[i]}', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

### **Evaluación estratégica de las tiendas de Alura Store**

**Analista:** Martinez Agustin Francisco.

##**Introducción**

El Sr. Juan, propietario de la cadena Alura Store, ha manifestado la necesidad de vender una de sus cuatro tiendas con el objetivo de obtener capital para iniciar un nuevo emprendimiento. Para tomar una decisión estratégica basada en datos, se realizó un análisis comparativo del desempeño de las cuatro sucursales.

Se evalúa informacion de los ingresos totales, las categorias mas vendidas, las reseñas de los clientes y el envio promedio por cada sucursal, con el fin de identificar la tienda menos eficiente.

##**Análisis**

###**Ingresos totales**
El análisis de facturación evidencia diferencias claras entre las sucursales. La Tienda 1 se posiciona como la de mayor nivel de ingresos dentro de la cadena, seguida por la Tienda 2. En tercer lugar se encuentra la Tienda 3, mientras que la Tienda 4 registra el menor volumen de facturación. Esta última presenta la contribución más baja al ingreso global de Alura Store, lo que reduce su peso dentro del desempeño financiero general.


###**Calificación promedio de clientes**

En relación con la satisfacción del cliente, la Tienda 3 obtiene la mejor calificación promedio, lo que indica una percepción positiva por parte de los consumidores. La Tienda 2 también mantiene una valoración alta y estable. La Tienda 4 presenta una calificación adecuada, aunque sin destacarse frente a las demás. Por su parte, la Tienda 1 muestra la valoración promedio más baja del conjunto, lo que sugiere oportunidades de mejora en la experiencia del cliente.

### **Costos promedio de envío**

El análisis del costo promedio de envío muestra que la Tienda 1 presenta el valor más alto, seguida por la Tienda 2. La Tienda 3 mantiene un costo intermedio, mientras que la Tienda 4 registra el costo promedio más bajo. Si bien este último aspecto representa una ventaja competitiva en términos logísticos, dicha condición no se refleja en un mayor nivel de ingresos ni en una mejora sustancial en su posicionamiento relativo.

### **Distribución de ventas por categoría**

En cuanto a la composición de ventas por categoría, las cuatro tiendas presentan estructuras relativamente equilibradas entre muebles, electrónicos, juguetes, electrodomésticos, deportes y diversión, libros, instrumentos musicales y artículos para el hogar. No se observan dependencias excesivas en una única categoría que representen un riesgo significativo para alguna sucursal. En términos generales, la distribución comercial es homogénea, por lo que no constituye un factor determinante para diferenciar el desempeño global entre las tiendas.

## **Evaluación estratégica**

Al considerar conjuntamente los indicadores analizados, se observa que:

- La **Tienda 1** lidera en ingresos, aunque presenta oportunidades de mejora en satisfacción del cliente.

- La **Tienda 2** muestra un desempeño sólido y equilibrado.

- La **Tienda 3** combina buen nivel de ingresos con la mejor percepción por parte de los clientes.

- La **Tienda 4** registra el menor nivel de ingresos y no presenta ventajas competitivas suficientes que compensen esta diferencia.

Si bien la Tienda 4 tiene el menor costo de envío, dicha ventaja no impacta de forma significativa en el volumen de ventas ni en su posicionamiento general dentro del grupo.

## **Conclusión**

Recomendaria la venta de la **Tienda 4** debido a que:

- Tiene la menor contribución a los ingresos totales.

- No lidera en satisfacción del cliente.

- No presenta una ventaja competitiva diferencial que justifique su mantenimiento dentro de la cadena Alura Store.

- Su desempeño global es inferior en comparación con las demás sucursales.

Su venta representa la alternativa más coherente para financiar el nuevo emprendimiento del Sr. Juan, minimizando el impacto negativo en el rendimiento global de la empresa.
